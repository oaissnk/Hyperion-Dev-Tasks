```python
import pandas as pd
```

```python
df = pd.read_csv('balance.txt',sep=' ')
df.head()
```

Write the code needed to produce a report that provides the following information:
* Compare the average income based on ethnicity.
* On average, do married or single people have a higher balance?
* What is the highest income in our dataset?
* What is the lowest income in our dataset?
* How many cards do we have recorded in our dataset? (Hint: use `sum()`)
* How many females do we have information for vs how many males? (Hint: use `count()` for a list of all methods for computation of descriptive stats, explore the [pandas documentation](https://pandas.pydata.org/pandas-docs/stable/reference/frame.html#computations-descriptive-stats)).




```python
average_income_eth = df.groupby('Ethnicity')['Income'].mean()
average_income_eth
```

```python
higher_balance = df.groupby('Married')['Balance'].mean()

higher_balance
#No is 13.49 which is > than Yes which is 13.38 therefore single people on average have a higher balance. 
```

```python
highest_row_income = df.sort_values('Income', ascending=False).iloc[0]
highest_row_income
#could use the max function but this way is better for series representation. Answer is 186.634
```

```python
lowest_row_income = df.sort_values('Income', ascending=True).iloc[0]
lowest_row_income
#could use min like the previous question. Answer is 10.354. Default for ascending is True anyway but added it in for visual clarity
```

```python
total_cards = df['Cards'].sum()
total_cards
#answer is 1183
```

```python
num_females = df[df['Gender']=='Female'].Gender.count() 
num_males = df[df['Gender']=='Male'].Gender.count()
print(f"number of females {num_females}")
print(f"number of females {num_males}")
```
